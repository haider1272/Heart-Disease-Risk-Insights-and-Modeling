### BMI vs Fast Food Consumption
<div style="background-color:#87cefa; padding: 20px; border-radius: 12px; text-align: center; margin-bottom: 20px;">
How does **BMI vary with weekly fast food consumption**, and do individuals who eat fast food more often show the **highest BMI values**?
</div>


```python
bmi_avg = df.groupby('Fast_Food_Meals_Per_Week', observed=True)['BMI'].mean()
bmi_avg
```

```python
fig = px.scatter(
    df,
    x='Fast_Food_Meals_Per_Week',
    y='BMI',
    trendline='ols',
    title='BMI vs Weekly Fast Food Consumption',
    labels={
        'Fast_Food_Meals_Per_Week': 'Weekly Fast Food Meals',
        'BMI': 'Body Mass Index'
    }
)
fig.show()
```

### Interpretation

The average BMI for different levels of weekly fast food consumption shows **some variation**, but a subtle trend can be observed:

- Individuals with **0‚Äì1 fast food meals per week** have an average BMI around **26.3‚Äì26.8**.  
- People eating **5‚Äì6 meals per week** show slightly higher BMI (**27.3‚Äì27.5**), suggesting that **frequent fast food intake may contribute to higher BMI**.  
- Interestingly, some higher meal counts (like 7‚Äì8 meals) show a small dip (~25.4‚Äì25.8), which could be due to **individual differences or smaller sample sizes** in these categories.  
- Overall, the data **supports a modest positive relationship** between weekly fast food consumption and BMI, aligning with the correlation and scatter plot analysis.  

This highlights the **potential impact of frequent fast food intake on body weight**, but also suggests that other lifestyle factors (physical activity, sleep, etc.) likely play a role.

## Digestive Issues vs Fast Food Consumption
<div style="background-color:#87cefa; padding: 20px; border-radius: 12px; text-align: center; margin-bottom: 20px;">
How do <strong>digestive issues</strong> vary with weekly fast food consumption, and do frequent consumers experience <strong>more digestive problems</strong>?
</div>


```python
digestive_counts = df.groupby('Fast_Food_Meals_Per_Week', observed=True)['Digestive_Issues'].value_counts(normalize=True).unstack()
digestive_counts
```

```python
digestive_plot = digestive_counts.reset_index().melt(
    id_vars='Fast_Food_Meals_Per_Week',
    value_name='Proportion',
    var_name='Digestive Issues'
)

fig = px.bar(
    digestive_plot,
    x='Fast_Food_Meals_Per_Week',
    y='Proportion',
    color='Digestive Issues',
    title='Digestive Issues vs Weekly Fast Food Consumption',
    labels={'Fast_Food_Meals_Per_Week': 'Weekly Fast Food Meals'},
    text_auto='.2f'
)
fig.show()
```

### Interpretation

The plot shows how <strong>digestive issues</strong> vary with weekly fast food consumption:

- Individuals with **0‚Äì1 fast food meals per week** report about **37‚Äì40%** experiencing digestive issues.  
- People consuming **2‚Äì7 meals per week** show slightly higher digestive problems (~38‚Äì47%).  
- Those with **very high fast food intake (10‚Äì14 meals/week)** have slightly lower reported digestive issues (~31‚Äì42%), which could be due to **smaller sample sizes or individual differences**.  

Overall, the data indicates a **moderate link between frequent fast food consumption and digestive discomfort**, emphasizing that diet may affect digestive health, but other lifestyle factors may also play a role.

## Digestive Issues vs BMI
<div style="background-color:#87cefa; padding: 20px; border-radius: 12px; text-align: center; margin-bottom: 20px;">
How do <strong>digestive issues</strong> vary with <strong>BMI</strong>, and do individuals with higher BMI experience more digestive problems?
</div>


```python
bins = [0, 18.5, 24.9, 29.9, 100]
labels = ['Underweight', 'Normal', 'Overweight', 'Obese']
df['BMI_Group'] = pd.cut(df['BMI'], bins=bins, labels=labels)

digestive_bmi = df.groupby('BMI_Group', observed=True)['Digestive_Issues'].value_counts(normalize=True).unstack()
digestive_bmi
```

```python
digestive_bmi_reset = digestive_bmi.reset_index().melt(id_vars='BMI_Group', value_vars=['No', 'Yes'],
                                                      var_name='Digestive_Issues', value_name='Proportion')

fig = px.bar(
    digestive_bmi_reset,
    x='BMI_Group',
    y='Proportion',
    color='Digestive_Issues',
    barmode='stack',
    title='Digestive Issues vs BMI Group',
    labels={'BMI_Group':'BMI Group', 'Proportion':'Proportion'}
)
fig.show()
```

### Interpretation

The data shows how <strong>digestive issues</strong> vary across different <strong>BMI groups</strong>:

- **Underweight** individuals have the **lowest proportion of digestive issues (~24%)**.  
- **Normal** and **Overweight** groups show a **similar proportion (~38‚Äì39%)**.  
- **Obese** individuals have a **slightly lower proportion (~38%)** than Overweight, but still higher than Underweight.  

**Insight:** Digestive issues are generally **more common in individuals with normal or higher BMI**, while being underweight is associated with fewer digestive problems. The trend is not strictly linear, suggesting other factors may influence digestive health.

## Health Variables Correlation
<div style="background-color:#87cefa; padding: 20px; border-radius: 12px; text-align: center; margin-bottom: 20px;">
How are the various health-related variables such as <strong>Age</strong>, <strong>BMI</strong>, <strong>Fast Food Meals per Week</strong>, <strong>Sleep Hours</strong>, and <strong>Energy Level</strong> correlated with each other, and which pairs show the strongest relationships?
</div>


```python
numerical_df = df.select_dtypes(include='number')
correlation_matrix = numerical_df.corr()
correlation_matrix
```

```python

plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix of Numerical Columns")
plt.show()
```

### Correlation Interpretation

- Most correlations between variables are **very weak**, with values close to 0. This suggests that in this dataset, these factors do not strongly predict each other.  
- **Age** has almost no correlation with any variable (all values between -0.07 and 0.05).  
- **Fast food meals per week** shows almost no correlation with BMI, energy levels, or overall health.  
- **BMI** has a very slight positive correlation with **Overall Health Score** (0.061) and **Average Daily Calories** (0.048), but these are negligible.  
- **Physical activity** shows a minor positive correlation with **Energy Level Score** (0.059), suggesting slightly higher activity may be associated with more energy.  
- **Sleep Hours** has weak negative correlations with **Energy Level Score** (-0.072) and **Overall Health Score** (-0.060), but effects are minimal.  
- **Doctor Visits** are essentially uncorrelated with all other variables (all near 0), meaning visit frequency does not seem strongly linked to health indicators in this dataset.  

**Insight:** Overall, the dataset does not show strong linear relationships between these health-related variables. Most lifestyle factors have minimal direct correlation with BMI, energy, or overall health in this sample. This highlights that health outcomes may depend on **a combination of many subtle factors** rather than any single variable.
